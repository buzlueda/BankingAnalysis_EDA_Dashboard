# Banking Dashboard – End-to-End EDA Notebook

## 1️⃣ Problem Statement:
Banks and financial institutions face the constant challenge of minimizing financial risk while providing loans to customers. The main problem is how to assess the likelihood of a customer repaying a loan and making data-driven decisions to reduce potential losses.

##2️⃣ Solution:
Using dashboards created with Power BI, the bank can visualize and analyze customer profiles efficiently.

The solution integrates multiple customer and banking tables to provide insights such as:

* Which customers are likely to repay loans
* How advisors’ performance relates to customer portfolios
* Distribution of income, assets, and risk profiles across the client base

Decision-makers can approve or reject loan applications based on data, not just intuition, improving efficiency and reducing financial risk.

## 3️⃣ About Dataset:

The dataset consists of multiple interlinked tables representing different aspects of customers and their banking relationships.

* Client-Banking Table: Contains information about each client’s financial accounts, loans, deposits, and engagement timeframe.

* Banking Relationship Table: Captures customer types (Retail, Institutional, Private Bank, Commercial), interactions with the bank, including advisor assignments and risk weightings.

* Gender Table: Contains gender information for demographic analysis.

* Investment Advisor Table: Details the advisors associated with clients, allowing for advisor-based performance insights.

Together, these tables allow for end-to-end analysis, from basic exploratory data analysis to visualizations and risk assessment dashboards.

**4.4** Feature Engineering

```python
# Define income band boundaries
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

# Create the 'Income Band' column using pd.cut
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
ax = df['Income Band'].value_counts().plot(kind='bar', figsize=(8,5))

for container in ax.containers:
    ax.bar_label(container, label_type='edge')

plt.title("Income Band Distribution")
plt.xlabel("Income Band")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
numerical_cols = ['Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
                  'Foreign Currency Account', 'Business Lending', 'Properties Owned']

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='rocket', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
pairs_to_plot = [
    ('Bank Deposits', 'Saving Accounts'),
    ('Checking Accounts', 'Saving Accounts'),
    ('Checking Accounts', 'Foreign Currency Account'),
    ('Age', 'Superannuation Savings'),
    ('Estimated Income', 'Checking Accounts'),
    ('Bank Loans', 'Credit Card Balance'),
    ('Business Lending', 'Bank Loans'),
]

for x_col, y_col in pairs_to_plot:
    plt.figure(figsize=(5, 3))
    sns.regplot(
        data=df,
        x=x_col,
        y=y_col,
        scatter_kws={'alpha': 0.4},
        line_kws={'color': 'red'}
    )
    plt.title(f'Relationship between {x_col} and {y_col}', fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.tight_layout()
    plt.show()
```

## Insights:

*   **Deposits and Savings Patterns:**
The strong correlation observed between Bank Deposits and Saving Accounts indicates that these metrics may capture overlapping aspects of customer financial behavior. Customers who actively deposit funds often maintain or grow their savings, highlighting consistent banking engagement and disciplined money management.
*   **Income, Age, and Wealth Accumulation:**
Moderate correlations between Age, Estimated Income, and account balances (including Superannuation, Savings, and Checking Accounts) reflect typical financial lifecycle trends. Generally, older customers and higher-income earners tend to accumulate more savings and retirement funds, and may carry higher balances on credit cards or loans.
* **Property Ownership and Limited Correlation:**
Ownership of properties appears less connected to banking variables. Factors such as location, real estate market conditions, or inheritance likely drive property acquisition. This explains the relatively weak correlations with deposits, savings, or loans.
* **Business vs. Personal Banking Behavior:**
Moderate correlations between Business Lending and Bank Loans suggest that some customers manage both personal and business financial obligations. However, the low correlation between business lending and deposit or property metrics indicates that business loans may cater to a distinct segment of customers or specific financial needs.
